```python
finance["month"]  = pandas.to_datetime(finance["Month"]).dt.month_name()
finance["year"]  = pandas.to_datetime(finance["Month"]).dt.year
```

```python
finance["Total_profit"] = finance["Revenue"]-finance["COGS"]
finance
```

```python
profit_trend_yearly = finance.groupby("year")[["Revenue","Total_profit"]].sum()
profit_trend_yearly
```

```python
profit_trend_yearly.plot(kind="bar",figsize=(10,6))
plt.xlabel("Year")
plt.ylabel("Profit")
plt.title("Profit Trend Yearly")
plt.legend(["Revenue","Total Profit"])
plt.tight_layout()
plt.show()
```

```python
profit_trend_monthly = finance.groupby("month")[["Revenue","Total_profit"]].sum().reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
profit_trend_monthly
```

```python
ax = profit_trend_monthly.plot(kind="line",figsize=(10,6))
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Profit Trend Monthly")
plt.legend(["Revenue","Total Profit"])
plt.xticks(range(len(profit_trend_monthly.index)),profit_trend_monthly.index)
plt.tight_layout()
plt.show()
```

```python
product_performance_revenue = finance.groupby("Product/Service")["Revenue"].sum().sort_values(ascending=False)
product_performance_revenue
```

```python
product_performance_revenue.plot(kind="bar",figsize=(10,6),color = ["green","blue","cyan","orange","red"])
plt.xlabel("Product/Service")
plt.ylabel("Revenue")
plt.title("Product Performance Revenue")
plt.tight_layout()
plt.show()
```

```python
product_performance_profit = finance.groupby("Product/Service")["Total_profit"].sum().sort_values(ascending=False)
product_performance_profit
```

```python
product_performance_profit.plot(kind="bar",figsize=(10,6),color = ["green","blue","cyan","orange","red"])
plt.xlabel("Product/Service")
plt.ylabel("Profit")
plt.title("Product Performance Profit")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
def aging_bucket(days):
  if days <= 30:
    return "0-30"
  elif days <= 60:
    return "31-60"
  elif days <=90:
    return "61-90"
  else:
    return "90+"
finance["aging_bucket"] = finance["Receivables Aging (Days)"].apply(aging_bucket)
```

```python
aging_summary = finance.groupby("aging_bucket")["Revenue"].sum()
aging_summary
```

```python
plt.figure(figsize=(10,6))
plt.bar(aging_summary.index,aging_summary.values/1e6)
plt.xlabel("Aging Bucket")
plt.ylabel("Revenue (millions)")
plt.title("Revenue by Aging Bucket")
plt.show()
```

```python
finance["net_cash"] = finance["Cash Inflows"] - finance["Cash Outflows"]
finance
```

```python
waterfall = finance.groupby("year")["net_cash"].sum()
waterfall
```

```python
plt.figure(figsize=(10,6))
plt.pie(waterfall.values,labels=waterfall.index,autopct="%1.1f%%")
plt.xlabel("Year")
plt.ylabel("Net Cash")
plt.title("Net Cash Flow by Year")
plt.show()
```

1.Objective:-
The objective of this dashboard is to analyze the financial performance of the business using key financial metrics such as Revenue, Profitability, Budget Variance, Receivables Aging, and Cash Flow.

2.Key Visualizations Implemented:-
Budget vs Actual Revenue Compares planned revenue with actual revenue to identify performance gaps.
Product/Service Performance Highlights which products generate the most revenue and profit.
Receivables Aging Categorizes outstanding payments into aging groups (0–30, 31–60) to monitor payment delays.
Cash Flow Waterfall Shows how cash moves through inflows and outflows to determine overall net cash position.

4.Key Insights:-
The dashboard reveals revenue trends over time and identifies top-performing products and regions. Budget variance helps determine whether the business is meeting financial targets.
Cash flow analysis helps track how operational activities affect the company’s financial health.

5.Assumptions:-
Revenue values represent actual sales for each period.
Cash inflows and outflows reflect operational financial movements.
Receivables aging represents the number of days invoices remain unpaid.
Budget values represent the planned financial targets set by the business.
